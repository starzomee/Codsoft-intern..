```python
import sklearn
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
full_data['scaled_time'] = scaler.fit_transform(full_data['Time'].values.reshape(-1, 1))
```

```python
full_data['scaled_amount'] = scaler.fit_transform(full_data['Amount'].values.reshape(-1,1))
```

```python
# We can drop the original Time and Amount columns
full_data.drop(['Time', 'Amount'], axis=1, inplace=True)
```

```python
# We can also rearrange the columns so that the Class column is the last one
full_data = full_data[[c for c in full_data if c not in ['Class']] + ['Class']]
```

```python
# Check for duplicates
full_data.duplicated().sum()
# There are 1081 duplicates
```

```python
# There are 1081 duplicates
# We can drop them to avoid overfitting
full_data.drop_duplicates(inplace=True)
```

# "Data Manipulation"

```python
# Create new features
# We can create a new feature that represents the ratio of the scaled amount to the scaled time
full_data['amount_time_ratio'] = full_data['scaled_amount'] / full_data['scaled_time']
```

```python
# Transform existing features
# We can apply a logarithmic transformation to the scaled amount feature to reduce skewness
full_data['log_scaled_amount'] = np.log(full_data['scaled_amount'] + 1)
```

```python
# Select relevant features
# We can use a correlation matrix to see the relationship between the features and the target variable
plt.figure(figsize=(12,10))
sns.heatmap(full_data.corr(), cmap='coolwarm')
plt.show()
```
